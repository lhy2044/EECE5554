```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

file = './imu_b.csv'

data = pd.read_csv(file)
df = pd.DataFrame(data)
time = df['timestamp'] - df['timestamp'][0]
vectornav_string = df["data"]
```

```python
euler_x = []
euler_y = []
euler_z = []
mag_x = []
mag_y = []
mag_z = []
acc_x = []
acc_y = []
acc_z = []
gyro_x = []
gyro_y = []
gyro_z = []

# split the string into a list of strings
for i in range(len(vectornav_string)):
    string = vectornav_string[i].split(',')
    euler_x.append(float(string[1]))
    euler_y.append(float(string[2]))
    euler_z.append(float(string[3]))
    mag_x.append(float(string[4]))
    mag_y.append(float(string[5]))
    mag_z.append(float(string[6]))
    acc_x.append(float(string[7]))
    acc_y.append(float(string[8]))
    acc_z.append(float(string[9]))
    gyro_x.append(float(string[10]))
    gyro_y.append(float(string[11]))
    gyro_z.append(float(string[12].split('*')[0]))
    
gyro_x = np.array(gyro_x)
gyro_y = np.array(gyro_y)
gyro_z = np.array(gyro_z)
acc_x = np.array(acc_x)
acc_y = np.array(acc_y)
acc_z = np.array(acc_z)
mag_x = np.array(mag_x)
mag_y = np.array(mag_y)
mag_z = np.array(mag_z)
euler_x = np.array(euler_x)
euler_y = np.array(euler_y)
euler_z = np.array(euler_z)
```

```python

import allantools

# Gyro X
(tau, adev, adev_error, adev_n) = allantools.oadev(gyro_x, rate=1, data_type="freq")
plt.loglog(tau, adev, label='Gyro X')
plt.loglog(tau, adev_error, label='Gyro X Error')
plt.xlabel('tau')
plt.ylabel('Allan Deviation')
plt.title('Gyro X Allan Deviation')
plt.legend()
plt.show()

B_gyro_x = adev[1]
N_gyro_x = adev[tau == 1]
K_gyro_x = (adev[2] - adev[1]) / (tau[2] - tau[1])

print(f"Gyro X: B = {B_gyro_x}, N = {N_gyro_x}, K = {K_gyro_x}")
```

```python
# Gyro Y
(tau, adev, adev_error, adev_n) = allantools.oadev(gyro_y, rate=1, data_type="freq")
plt.loglog(tau, adev, label='Gyro Y')
plt.loglog(tau, adev_error, label='Gyro Y Error')
plt.xlabel('tau')
plt.ylabel('Allan Deviation')
plt.title('Gyro Y Allan Deviation')
plt.legend()
plt.show()

B_gyro_y = adev[1]
N_gyro_y = adev[tau == 1]
K_gyro_y = (adev[2] - adev[1]) / (tau[2] - tau[1])

print(f"Gyro Y: B = {B_gyro_y}, N = {N_gyro_y}, K = {K_gyro_y}")
```
